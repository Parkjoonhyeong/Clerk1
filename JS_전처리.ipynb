```python
columns = ['CWE-79', 'CWE-89', 'CWE-352', 'CWE-78']

# 'CWE ID' 컬럼 값이 columns 리스트에 포함된 행만 필터링하여 새로운 데이터프레임 생성
filtered_dataset = df[df['CWE ID'].isin(columns)].reset_index(drop=True)
```

```python
count = filtered_dataset.groupby(['CWE ID', 'vul']).size()
count_sorted = count.sort_values(ascending=False)
count_sorted
```

```python
# 'CWE ID'가 'CWE-79' 또는 'CWE-89'인 행만 선택
filtered_df = df[df['CWE ID'].isin(['CWE-79', 'CWE-89','CWE-352','CWE-78'])]

# 각 'CWE ID'의 최대 500개 행만 남기기
cwe_79_rows = filtered_df[filtered_df['CWE ID'] == 'CWE-79'].head(500)
cwe_89_rows = filtered_df[filtered_df['CWE ID'] == 'CWE-89'].head(500)
cwe_352_rows = filtered_df[filtered_df['CWE ID'] == 'CWE-352'].head(500)
cwe_78_rows = filtered_df[filtered_df['CWE ID'] == 'CWE-78'].head(1500)

# 필요한 행들을 다시 합치기
final_df = pd.concat([cwe_79_rows, cwe_89_rows,cwe_352_rows]).reset_index(drop=True)
```

```python
import re

def preprocess_text(text):
    # 주석 제거
    text_without_comments = re.sub(r'\/\/.*|\/\*[\s\S]*?\*\/', '', text)

    # 줄바꿈, 탭 등 공백 문자 제거
    cleaned_text = re.sub(r'[\n\r\t\s]', ' ', text_without_comments)

    return cleaned_text

# 'vul' 열의 문자열 데이터 전처리 함수 적용
final_df['code'] = final_df['code'].apply(preprocess_text)

# 전처리된 결과 확인
print("전처리 전 데이터프레임:")
final_df

# 'vul_cleaned' 열만 선택하여 최종 전처리된 데이터프레임 출력
final_processed_df = final_df[['CWE ID', 'code']]
print("\n전처리 후 데이터프레임:")
final_processed_df
```
